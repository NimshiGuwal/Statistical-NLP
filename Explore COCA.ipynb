```python
import nltk
```

```python
#Task 1: Extract text from corpus
# IMPORT INAUGURAL CORPUS
from nltk.corpus import inaugural
#DIFFERENT CATEGORIES IN REUTERS CORPUS
inaugural.fileids()
```

```python
texts=""
texts=inaugural.words(fileids='1993-Clinton.txt')[:200]
texts
```

```python
# Task 2:Tokenize them
#tokenizing the a long paragrah to sentences and words from clinton.txt
from nltk.tokenize import sent_tokenize
#converting list to single string str1
str1=" ".join(texts)
tokenized_text=sent_tokenize(str1)
print(tokenized_text)
```

```python
#Task 3:Pos_tagging
#parts of speech tagging
tagged = nltk.pos_tag(str1)
print(tagged)
```

```python
#Task 4: identifying stems
#using port stemmer to reduce the words to their rootword and remove any derivational affixes
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
print("Orignal Sentence:",str1)
print("\n")

str1 = [stemmer.stem(token) for token in str1.split(" ")]
print("Stemmed Sentence:")
print(" ".join(str1))
```
