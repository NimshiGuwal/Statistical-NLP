```python
#import spacy package
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
example = "Google, a company founded by Larry Page and Sergey Brin in the United States of America "\
+ "has one of the world’s most advanced search engines."
```

```python
doc = nlp(example)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
from spacy import displacy
displacy.render(doc, style='ent')
```

```python
text = """Business area Sandvik Mining and Rock Technology has partnered with Nokia to develop solutions for private LTE and 5G technology at the company’s test mine in Tampere, Finland. Hardware at the mine is already 5G capable and will be upgraded with 5G software as soon as it is ready"""
```

```python
import nltk
#tokens
tokens = nltk.word_tokenize(text)
```

```python
#tagging
pos_tagggs1 = nltk.pos_tag(tokens)
```

```python
print(pos_tagggs1[:100])
```

```python
#Define Chunk Rule
#grammar = "CHUNK: {<DT><JJ>*<NN><VBD>?<DT>?}"
grammar = "NP: {<DT>?<JJ>*<NN>}"
#Using above defined grammer, created chunk parser
cp = nltk.RegexpParser(grammar)
#apply parser on pos tags
result = cp.parse(pos_tagggs1)
```

```python
print(result)
```

```python
result.draw()
```

```python
import spacy
for ent in text.ents:
    print(ent.result, ent.label_, spacy.explain(ent.label_))
```
